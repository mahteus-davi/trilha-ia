# Analise Bivariada

```python
# Hipotese? Clientes com contratos do tipo mensal sao mais propensos ao Churn
# Construir tabelas de contingencia entre tipo de contrato e Churn - Quantidade
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='total')
```

```python
# Construir tabela de contingencia entre tipo de contrato e Churn - Proporcao
pd.crosstab(df_churn.Churn, df_churn.Contract, normalize='index', margins=True, margins_name='total')
```

Considerando o publico total, os clientes de contrato mensal representam 55%. Porem, se considerarmos apenas os clientes que abandonaram o servico, 88% tinham contrato mensal. Dessa forma, podemos afirmar que ha uma correlacao entre o tipo de contrato mensal e o abadono.

```python
# Avaliar a correlacao entre duas variaveis categoricas (qualitativas)
# Executar um teste de hipotese chamado  Chi-Square ou Qui-Quadrado de Pearson
# Num teste de hipotese, duas hipoteses sao formuladas:
# H0 (Hipotese Nula): as duas variaveis sao independentes
# H1 (Hipotese Complementar): as duas variaveis nao sao independentes

# O teste serve para confirmar ou recusar a hipotese nula (H0)
# Quando a probabilidade de observarmos H0 e inferior a 0.05 (p-value), recusamos a hipote nula e seguimos com a complementar.
```

```python
# gerar um DF da Crosstab (sem totais)
df_cosstab_churn_contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
```

```python
# Calcular os Scores e P-Values
from scipy.stats import chi2_contingency
```

```python
# Tupla = Lista imutavel
chi_scores_churn_contract = chi2_contingency(df_cosstab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
score_churn_contract = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
# Apresentar numeros com decimais sem a notacao cientifica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
# Criar DataFrame com os resultados
df_chi_score_churn_contract = pd.DataFrame({'Qui2': score_churn_contract, 'p-value': pvalues_churn_contract})
```

```python
df_chi_score_churn_contract
```

Conforme o teste estatistico (Chi-Square), o P-value <= 0.05, desta forma rejeitamos a hipotese nula, ou seja, as variaveis nao sao independentes. Pelo Qui2 alto, podemos afirmar que ha uma forte correlacao.

```python
# Avaliar a correlacao entre uma variavel qualitativa e quantitativa
# usar o mesmo teste de Chi-Square, mas transformar a variavel quantitativa em qualitativa
# Hipotese: Cliente com menos de 6 meses de contrato e mais propenso ao churn

# Criar uma coluna nova no DataFrame com base numa condicao de outra coluna
# Coluna = 'TempoMenor6Meses'

import numpy as np
```

```python
# Criar uma nova coluna
df_churn['TempoMenor6Meses'] = np.where(df_churn.tenure<6, 'Yes', 'No')
```

```python
# Checar o DataFrame
df_churn.head(5)
```

```python
# Gerar um DF da Crosstab (sem totais)
df_cosstab_churn_tenure = pd.crosstab(df_churn.Churn, df_churn.TempoMenor6Meses)
```

```python
# Calcular os Scores
chi_scores_churn_tenure = chi2_contingency(df_cosstab_churn_tenure)
```

```python
scores_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
# Criar DataFrame com os resultados
df_chi_score_churn_tenure = pd.DataFrame({'Qui2': scores_churn_tenure, 'p-values': pvalues_churn_tenure})
```

```python
# Analisar os resultados
df_chi_score_churn_tenure
```

Como o P-Value <= 0.05, rejeitamos a H0, ou seja, as variaveis nao sao independentes. E vale mencionar que com base do Qui2 resultante das 2 analises, a correlacao entre Churn e Tempo de contrato < 6 meses e menor do que a correlacao entre Churn e Tipo de contrato

```python
# Correlacao entre 2 variaveis numericas
# Tenure com TotalCharges
# A intuicao e que quanto mais tempo de contrato maior o valor pago.

# Correlacao entre 2 variaveis numericas - Pearson

df_churn.tenure.corr(df_churn.TotalCharges)
```

```python
# Criar um DataFrame somente com clientes do tipo Month to Month (Contrato Mensal)
df_churn_month = df_churn[df_churn.Contract=='Month-to-month']
```

```python
df_churn_month
```

```python
# Histograma - Checar visualmente se os dados seguem uma distribuicao normal
df_churn_month.TotalCharges.plot.hist()
```

```python
# Metodo de Tukey - IQR (Distribuicao nao normal)
# IQR - Range Interquartil
# IQR = 3o quartil - 1o quartil
q1_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.25)
q3_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.75)
iqr_TotalCharges_month = q3_TotalCharges_month - q1_TotalCharges_month
iqr_TotalCharges_month
```

```python
# Limites Inferior e Superior

limInf_TotalCharges_month = q1_TotalCharges_month - (iqr_TotalCharges_month * 1.5)
limSup_TotalCharges_month = q3_TotalCharges_month + (iqr_TotalCharges_month * 1.5)
limInf_TotalCharges_month, limSup_TotalCharges_month
```

```python
df_churn_month[(df_churn_month.TotalCharges < limInf_TotalCharges_month) | (df_churn_month.TotalCharges > limSup_TotalCharges_month)]
```

```python
# ZScore (Usando para distribuicao Normal)
# Indica quantos desvios padroes um ponto especifico de dados esta distante da Media
# zscore = (x - media) / desvio padrao

from scipy.stats import zscore
```

```python
z = np.abs(zscore(df_churn_month.TotalCharges))
```

```python
z
```

```python
df_churn_month[z > 3.0]
```
